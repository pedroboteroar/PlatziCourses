# Matplotlib

Pyplot es una herramienta que tiene matplotlib para ejecutar y crea gráficas de manera sencilla.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11) # Crea un arreglo de 11 elementos entre 0 y 5
y = x ** 2
```

```python
plt.plot(x, y, 'ro') # Grafica x vs y con color rojo en forma de puntos
plt.show() # Muestra el gráfico
```

```python
plt.plot(x, y, 'bs-') # Grafica x vs y con color azul en forma de línea
plt.show() # Muestra el gráfico
```

```python
plt.plot(x, y, 'yD:') # Grafica x vs y con color amarillo en forma de diamantes
plt.show() # Muestra el gráfico
```

Si queremos ggraficar un histograma podemos hacerlo de la siguiente manera,

```python
plt.hist(x)
plt.show()
```

Para un boxplot, solo debemos cambiar el llamado de la grafica

```python
plt.boxplot(x)
plt.show()
```

Subplot me permite crear mas de un solo grafico al mismo tiempo. En este caso, creamos una grafica de 1 fila y ods columnas que incluya un plot y un histograma

```python
plt.subplot(1,2,1) # 1 fila, 2 columnas, posición 1
plt.plot(x, y)
plt.subplot(1,2,2) # 1 fila, 2 columnas, posición 2
plt.hist(y)
plt.show()
```

Lo que se modifica va a ser la grafica a la que estemos haciendo referencia, como se muestra en la siguiente grafica,

```python
plt.subplot(1,2,1) # 1 fila, 2 columnas, posición 1
plt.plot(x, y, 'r--')
plt.plot(y, x, 'g--')
plt.subplot(1,2,2) # 1 fila, 2 columnas, posición 2
plt.pie(y)
plt.show()
```

Pyplot es rápido, fácil e incluye una sola figura.

Object Oriented permite mas personalizacion, necesita mas codigo, es mas amigable para multiples diagramas y se tiene que un objeto define una figura, la cual actua como lienzo  para introducir diferentes axes que poseen sus propios ejes.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # [left, bottom, width, height]
axes.plot(x, y, 'b')
fig.show()
```

Si queremos mas, podemos hacer lo siguiente,

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) # [left, bottom, width, height]
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3]) # [left, bottom, width, height]

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')

axes2.set_facecolor('gray') # Cambia el color de fondo del gráfico para el axes2

fig.show()
```

Para trabajar multiples graficas en una sola en el metodo de orientado a objetos definimos lo siguiente,

```python
x = np.linspace(0,5,11) # Crea un arreglo de 11 elementos entre 0 y 5
y = np.sin(x)
```

```python
fig, axes = plt.subplots() # Crea una figura y un axes
axes.plot(x, y, 'b')
```

Del siguiente modo, podemos definir objetos y trabajr con ellos de manera independiente

```python
fig, axes = plt.subplots(nrows= 1, ncols= 2) # Crea una figura y un axes
axes[0].plot(x, y, 'b') # axes[0] es el primer gráfico
axes[1].plot(y, x, 'r') # axes[1] es el segundo gráfico
```

```python
fig, axes = plt.subplots(nrows= 2, ncols= 4) # Crea una figura y un axes
axes[0][0].plot(x, np.cos(x), 'b') # axes[0][0] es el primer gráfico
axes[0][1].plot(x, np.sin(x), 'r') # axes[0][1] es el segundo gráfico
axes[0][2].plot(x, np.tan(x), 'g') # axes[1][0] es el tercer gráfico
axes[0][3].plot(x, np.tanh(x), 'y') # axes[1][1] es el cuarto gráfico
axes[1][0].plot(x, np.cosh(x), 'c') # axes[1][0] es el quinto gráfico
axes[1][1].plot(x, np.sinh(x), 'm') # axes[1][1] es el sexto gráfico
axes[1][2].plot(x, np.arctan(x), 'k') # axes[1][2] es el séptimo gráfico
axes[1][3].plot(x, np.arctanh(x), 'tab:orange') # axes[1][3] es el octavo gráfico
fig.tight_layout() # Ajusta el espacio entre gráficos
```

Ahora, para agregar titulos, leyendas, etiquetas y demas debemos

```python
fig, axes = plt.subplots(nrows= 1, ncols= 2, figsize = (10, 5)) # Crea una figura y un axes
axes[0].plot(x, y, 'b', label= "$sin(x)$") # axes[0] es el primer gráfico
axes[0].set_title('Relacion X vs Y') # Título del gráfico 1
axes[0].set_xlabel('X') # Etiqueta del eje X
axes[0].set_ylabel('Y') # Etiqueta del eje Y
axes[0].legend() # Muestra la leyenda
axes[1].plot(y, x, 'r') # axes[1] es el segundo gráfico
axes[1].set_title('Relacion Y vs X') # Título del gráfico 2
axes[1].set_xlabel('Y') # Etiqueta del eje X
axes[1].set_ylabel('X') # Etiqueta del eje Y
fig.suptitle('Relacion entre Variables') # Título de la figura
fig.tight_layout() # Ajusta el espacio entre gráficos
fig.show()
```

Sin usar el metodo orientado a objetos,

```python
plt.plot(x,y, label= "sin(x)")
plt.title('Relacion X vs Y') # Título del gráfico 1
plt.xlabel('X') # Etiqueta del eje X
plt.ylabel('Y') # Etiqueta del eje Y
plt.legend() # Muestra la leyenda
plt.show()
```

Con pyplot podemos customizar los colores de manera que,

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax = plt.plot(x, x + 1, color = 'red', alpha = 0.8) # alpha es la transparencia, 0 es transparente y 1 es opaco
ax = plt.plot(x, x + 2, color = 'blue', linewidth = 5) # linewidth es el ancho de la línea, 1 es el valor por defecto
ax = plt.plot(x, x + 3, color = 'green', linestyle = '--') # linestyle es el estilo de la línea, '-' es el valor por defecto (línea continua
ax = plt.plot(x, x + 4, color = '#AA00AA', marker = 'o', markersize = 10) # marker es el marcador, 'None' es el valor por defecto (sin marcador). markersize es el tamaño del marcador, 6 es el valor por defecto
```

Un grafico de barras nos permite graficar variables categoricas

```python
country = ['USA', 'Canada', 'Germany', 'UK', 'France']
population = [325, 35, 82, 67, 65]
```

```python
plt.bar(country, population, width= 0.5, color= ['#2B3A67', '#496A81', '#66999B', '#B3AF8F', '#FFC482'], linewidth= 3)
plt.xticks(country, rotation= 90)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population by Country')
plt.show()
```

La misma grafica de manera horizontal seria:

```python
plt.barh(country, population, height= 0.5, color= ['#2B3A67', '#496A81', '#66999B', '#B3AF8F', '#FFC482'], linewidth= 3)
plt.yticks(country, rotation= 0)
plt.ylabel('Country')
plt.xlabel('Population')
plt.title('Population by Country')
plt.show()
```

```python
data = np.random.randint(0, 100, 100)
```

Para anlaizar distribucion y frecuencia es muy util el histograma

```python
plt.hist(data, bins= 10, color= '#2B3A67', edgecolor= '#496A81', linewidth= 3, histtype='bar') # bins es el número de barras que se quieren mostrar
# linewidth es el ancho de la línea que separa las barras
```

```python
plt.hist(data, bins= 10, color= '#2B3A67', edgecolor= '#496A81', linewidth= 3, histtype='step') # bins es el número de barras que se quieren mostrar
# linewidth es el ancho de la línea que separa las barras
```

El boxplot tambien os permite analizar la distribucion de los datos

```python
data = np.append(data, 200)
plt.boxplot(data, notch= True, vert= True, patch_artist= True, showfliers= True) # notch es para mostrar o no la muesca, vert es para mostrar la gráfica de forma vertical u horizontal, patch_artist es para rellenar la caja
# showfliers es para mostrar o no los outliers
plt.show()
```

Ahora, el grafica de dispersion o scatter es una grafica que nos permite ver la relacion entre dos variables, en este caso, la relacion entre el precio de las casas y el numero de habitaciones que tienen.

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2 
colors = np.random.rand(N) 
```

```python
plt.scatter(x, y, s= area, c= colors, alpha= 0.5) # s es el tamaño de los puntos, c es el color de los puntos, alpha es la transparencia de los puntos
plt.show()
```

```python
import seaborn as sns # Importa la librería seaborn
import pandas as pd # Importa la librería pandas
```

```python
sns.barplot(x= ['A', 'B', 'C', 'D', 'E'], y= [1, 2, 3, 4, 5]) # Crea un gráfico de barras
```

```python
sns.set(style= 'darkgrid', palette= 'mako', font= 'Verdana') # Cambia el estilo de la gráfica, la paleta de colores y la fuente
sns.barplot(x= ['A', 'B', 'C', 'D', 'E'], y= [1, 2, 3, 4, 5]) # Crea un gráfico de barras
plt.show()
```

```python
tips = sns.load_dataset('tips') # Carga el dataset tips de seaborn
```

```python
sns.displot(tips['total_bill'], bins= 20, kde= True) # Crea un histograma, kde es para mostrar o no la distribución de probabilidad
```

```python
sns.displot(data= tips, x= 'total_bill', y= 'tip') # Crea un gráfico de densidad
```

```python
sns.displot(data= tips, x= 'total_bill', y= 'tip', kind= 'kde') # Crea un gráfico de densidad
```

```python
sns.displot(data= tips, x= 'total_bill', y= 'tip', hue= 'sex', palette= 'mako') # Crea un gráfico de densidad con una variable categórica
```

```python
sns.displot(data= tips, x= 'total_bill', hue= 'sex', palette= 'mako') # Crea un gráfico de densidad con una variable categórica
```

```python
sns.displot(data= tips, x= 'total_bill', hue= 'sex', kind= 'kde', palette= 'mako') 
```

Para trabajar distribuciones, tenemos varias opciones como

```python
sns.histplot(data= tips, x= 'tip', bins= 20, kde= True) # Crea un histograma y una distribución de probabilidad
plt.show()
```

```python
sns.histplot(data= tips, x= 'tip', bins= 20, kde= True, cumulative= True) # Crea un histograma y una distribución de probabilidad acumulada
plt.show()
```

```python
sns.histplot(data= tips, x= 'tip', bins= 20, kde= True, hue= 'sex') # Crea un histograma y una distribución de probabilidad con una variable categórica
plt.show()
```

```python
sns.histplot(data= tips, x= 'tip', bins= 20, hue= 'sex', stat= 'count', multiple= 'stack') # Crea un histograma y una distribución de probabilidad con una variable categórica y con una estadística diferente
plt.show()
```

Otro grafico para la distribucion, es

```python
sns.kdeplot(data= tips, x= 'tip', hue= 'sex', palette= 'mako', cumulative= False, fill= True, bw_adjust= 1) # Crea una distribución de probabilidad
# bw_adjust es para ajustar la suavidad de la distribución, 1 es el valor por defecto
# shade es para mostrar o no la sombra de la distribución, False es el valor por defecto
plt.show()
```

Otro grafico es

```python
sns.ecdfplot(data= tips, x= 'tip', hue= 'sex', palette= 'mako') # Crea una función de distribución acumulada
plt.show()
```

```python
sns.displot(data= tips, x= 'tip', hue= 'sex', palette= 'mako', kind= 'kde')
```

Reto de la clase "Distribuciones"

```python
sns.displot(data=tips, x="tip", y="total_bill", kind="kde", hue="smoker", levels=4, palette= 'mako') # Crea un gráfico de densidad en 3D
```

Para trabajar datos categoricos, tenemos que podemos hacer las siguientes graficas

```python
sns.countplot(data= tips, x= 'day', hue= 'sex', palette= 'mako') # Crea un gráfico de barras con una variable categórica
plt.show()
```

```python
# Strip Plot
plt.figure(figsize= (10, 6))
sns.stripplot(data= tips, x= 'day', y= 'total_bill', hue= 'sex', dodge= True, palette= 'mako')
plt.show()
```

```python
# Swarm Plot
plt.figure(figsize= (10, 6))
sns.swarmplot(data= tips, x= 'day', y= 'total_bill', hue= 'sex', dodge= True, palette= 'mako')
plt.show()
```

```python
#boxplot
plt.figure(figsize= (10, 6))
sns.boxplot(data= tips, x= 'day', y= 'total_bill', hue= 'sex', dodge= True, palette= 'mako')
sns.swarmplot(data= tips, x= 'day', y= 'total_bill', hue= 'sex', dodge= True, palette= 'rocket')
plt.show()
```

```python
#boxplot
plt.figure(figsize= (10, 6))
sns.violinplot(data= tips, x= 'day', y= 'total_bill', hue= 'sex', split=True, dodge= True, palette= 'mako')
# split es para dividir la caja en dos
plt.show()
```

```python
# Cat Plot
sns.catplot(data= tips, x= 'day', y= 'total_bill', hue= 'sex', palette= 'mako', dodge= True, kind= 'box', col= 'smoker', row= 'time')
plt.show()
```

Para los graficos de tipo relacion, tenemos las siguientes graficas

```python
plt.figure(figsize= (10, 6))
sns.scatterplot(data= tips, x= 'total_bill', y= 'tip', hue= 'day', palette= 'mako', style= 'time', size= 'size', sizes= (20, 200), alpha= 0.7)
plt.legend(bbox_to_anchor= (1.05, 1), loc= 2, borderaxespad= 0) # bbox_to_anchor es para mover la leyenda, loc es para la posición de la leyenda
plt.show()
```

El siguiente grafico no me gusta mucho

```python
sns.lineplot(data= tips, x= 'total_bill', y= 'tip', hue= 'day', palette= 'mako', style= 'time', markers= True, dashes= False)
```

Por ultimo, relplot nos permite graficar cualquier tipo de grafico de relacion

```python
plt.figure(figsize= (10, 6))   
sns.relplot(data= tips, x= 'total_bill', y= 'tip', hue= 'day', palette= 'mako', style= 'time')
plt.show()
```

Otros dos graficos que nos permiten combinar tipos de graficos son jointplot y pairplot

```python
sns.jointplot(data= tips, x= 'total_bill', y= 'tip', hue= 'sex', palette= 'mako', kind= 'hist', marginal_ticks= True, 
              marginal_kws= dict(bins= 25, multiple= 'stack')) # Crea un gráfico de dispersión
plt.show()
```

Pairplot nos muestra la relacion que existe entre variables numericas

```python
sns.pairplot(data= tips, hue= 'sex', palette= 'mako') # Crea un gráfico de dispersión para cada par de variables numéricas
plt.show()
```

Por ultimo, vamos a ver el heatmap, que nos permite diagramar datos con estructura matricial

```python
corr = tips.corr() # Crea una matriz de correlación
```

```python
sns.heatmap(corr, annot= True, cmap= 'mako') # Crea un mapa de calor
```
