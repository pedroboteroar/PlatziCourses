Que es una función? 
- Una función es cómo una máquina donde entra un elemento x y sale un elemento y. En el caso de una variable $y = f(x)$.
- En una función, vemos que de una x no se pueden obtener más de dos valores y. 
- Se pueden representar de manera verbal, númerica, algebraica y visualmente.

Existen dos tipos de variables principales, los cuales son cualitativas y cuantitativas.


Cuantitativas
- Nominales - Clasificar
- Ordinales - Jerarquizar
- Binarias - Escoger


Cuantitativas
- Discretas - Contar
- Continuas - Medir

**Dominio de una función**

Los valores que toma $x$ y que estan definidos en la función $f(x)$.

**Rango de una función**

Todos los resultados que nos puede dar una función.

**Ejemplo**

Tenemos una máquina para hacer café.

El dominio son los granos de café.
La función es nuestra cafetera.
El rango son todas las clases de café que podemos preparar.

**Conjuntos**

- Los Diagramas de Venn son otra forma de representar un conjunto de manera gráfica. Los diagramas nos permiten comprender y analizar fácilmente los conjuntos.
- Conjunto vacio ($\phi$)

Nota: El conjunto que contenga el elemento

La unión de dos conjuntos A y B es el conjunto formado por todos los elementos de ambos conjuntos sin repetir ninguno y se representa así: 
$A ∪ B = \{x ∣ x ∈ A \lor x ∈ B\}$

La intersección de dos conjuntos A y B son el conjunto de elementos que forman parte tanto de A como de B, es decir: 
$A ∩ B = \{x ∣ x ∈ A \land x ∈ B\}$

La diferencia de dos conjuntos A y B, es un conjunto formado por todos los elementos de A que no están en B y se escribe así: 
$A / B = A - B = \{x ∈ A ∣ x ∉ B \}$

**Funciones lineales**

Tienen la forma de $f(x) = mx + b$ donde $m$ y $b$ $\in$ $\mathbb{R}$.

$m$ se puede calcular por $m = \frac{y_2 -  y_1}{x_2 - x_1}$ y $b$ es el punto de corte en el eje $y$. Du dominio es $Dom_f= (-\infty, \infty)$. Su imagen es $Im_f=(-\infty, \infty)$.

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x + b

x = np.linspace(-10, 10, N)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
plt.show()
```
