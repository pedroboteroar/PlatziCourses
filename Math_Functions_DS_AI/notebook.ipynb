Que es una función?

- Se pueden representar de manera verbal, númerica, algebraica y visualmente.

La unión de dos conjuntos A y B es el conjunto formado por todos los elementos de ambos conjuntos sin repetir ninguno y se representa así: 
$A ∪ B = \{x ∣ x ∈ A \lor x ∈ B\}$

La intersección de dos conjuntos A y B son el conjunto de elementos que forman parte tanto de A como de B, es decir: 
$A ∩ B = \{x ∣ x ∈ A \land x ∈ B\}$

La diferencia de dos conjuntos A y B, es un conjunto formado por todos los elementos de A que no están en B y se escribe así: 
$A / B = A - B = \{x ∈ A ∣ x ∉ B \}$

**Funciones lineales**

Tienen la forma de $f(x) = mx + b$ donde $m$ y $b$ $\in$ $\mathbb{R}$.

$m$ se puede calcular por $m = \frac{y_2 -  y_1}{x_2 - x_1}$ y $b$ es el punto de corte en el eje $y$. Du dominio es $Dom_f= (-\infty, \infty)$. Su imagen es $Im_f=(-\infty, \infty)$.

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x + b

x = np.linspace(-10, 10, N)
y = f(x)

graph(x,y)
plt.show()
```

**Funciones Polinómicas**

Tiene la forma de $f(x) = a_nx^n + a_{n-1}x^{n-1} + ... + a_1x + a_0$ donde $a_n, a_{n-1}, ..., a_1, a_0$ $\in$ $\mathbb{R}$ y $a_n \neq 0$.

Dando un ejemplo, tenemos que $f(x) = 2x^2 + 3x + 1$ donde $a_2 = 2$, $a_1 = 3$ y $a_0 = 1$.

Este ejemplo es una función polinómica de grado 2. 

El grado se mide por el exponente más grande de la función. En este caso es 2.

```python
# Polinomio Grado IMPAR
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

graph(x,y)
plt.show()
```

```python
# Polinomio Grado PAR
def f(x):
    return (2*x**8)-(x**4)+(3*x**2)+4

y = f(x)

graph(x,y)
plt.show()
```

**Funciones Potencia**

Tiene la forma de $f(x) = x^a$ donde $a$ $\in$ $\mathbb{R}$. 

Por ejemplo, tenemos que $f(x) = x^2$.

En este caso:

- El dominio es $Dom_f = (-\infty, \infty)$.
- La imagen es $Im_f = [0, \infty)$.

```python
# Ejemplo Funcion Potencia

def f(x):
    return x**8

y = f(x)

graph(x,y)
plt.show()
```

**Funciones Trascendentes**

Estas funciones no pueden ser expresadas como polinomios o la suma de polinomios.

Algunos ejemplos son:
- Función exponencial: $f(x) = e^x$.
- Función logarítmica: $f(x) = log(x)$.
- Función trigonométrica: $f(x) = sin(x)$.

```python
# Ejemplo Funcion Exponencial
def f(x):
    return np.exp(x)

y = f(x)

graph(x,y)
plt.show()
```

```python
# Ejemplo Funcion Logaritmica
def f(x):
    return np.log(x)

y = f(x)

graph(x,y)
plt.show()
```

```python
# Ejemplo Funcion Seno
def f(x):
    return np.sin(x)

y = f(x)

graph(x,y)
plt.show()
```

**Funciones Seccionadas**

Tiene la forma de $f(x) = \begin{cases} f_1(x) & x \in A \\ f_2(x) & x \in B \\ ... \\ f_n(x) & x \in C \end{cases}$

Como ejemplo, podemos ver la función escalón unitario $f(x) = \begin{cases} 0 & x < 0 \\ 1 & x \geq 0 \end{cases}$

```python
# Función Escalon
def f(x):
    return np.heaviside(x,1)

y = f(x)

graph(x,y)
plt.show()
```

**Funciones Compuestas**

Tiene la forma de $f(x) = f(g(x))$.

Como ejemplo, podemos ver la función $f(x) = e^{sin(x)}$.

Algunas propiedades de las funciones compuestas son:
- $f(g(x)) \neq g(f(x))$. 
- $f(g(x)) = g(f(x))$.

```python
# Ejemplo de Funcion Compuesta
# f(x) = e^(sin(x))
def f(x):
    return np.exp(np.sin(x))

y = f(x)

graph(x,y)
plt.show()
```

Ahora, viendo el ejemplo anterior, tenemos que $f(g(x)) = e^{sin(x)}$ y $g(f(x)) = sin(e^x)$.

**Funciones Inversas**

Tiene la forma de $f^{-1}(x)$.

Como ejemplo, podemos ver la función $f(x) = x^2$.

Para encontrar la función inversa, tenemos que despejar $x$ de la función $f(x)$.

$f(x) = x^2$

$x = \sqrt{f(x)}$

$f^{-1}(x) = \sqrt{x}$

```python
# Graficando f(x) y f-1(x)
def f(x):
    return x**2

def f_inv(x):
    return np.sqrt(x)

x = np.linspace(0, 10, N)
y = f(x)
y_inv = f_inv(x)

plt.subplots(1,2, figsize=(10,5))
plt.subplot(1,2,1)
graph(x,y)
plt.title('f(x)')
plt.subplot(1,2,2)
graph(x,y_inv)
plt.title('f_inv(x)')
plt.show()
```

Vemos que la función inversa es $f^{-1}(x) = \sqrt{x}$. Y que a nivel visual, la función inversa es la función original reflejada en la recta $y = x$.

**Funciones Racionales**

Tiene la forma de $f(x) = \frac{p(x)}{q(x)}$ donde $p(x)$ y $q(x)$ son polinomios.

Como ejemplo, podemos ver la función $f(x) = \frac{x^2 + 1}{x + 1}$.

Algunas propiedades de las funciones racionales son:
- $Dom_f = \{x \in \mathbb{R} \mid q(x) \neq 0\}$.
- $Im_f = \{y \in \mathbb{R} \mid y = \frac{p(x)}{q(x)}\}$.

```python
# Graficando la función racional
def f(x):
    return (x**2+1)/(x+1)

y = f(x)

graph(x,y)
plt.show()
```

**Como manipular funciones**

- Suma: $f(x) + g(x)$.
- Resta: $f(x) - g(x)$.
- Multiplicación: $f(x) \cdot g(x)$.
- División: $\frac{f(x)}{g(x)}$.
- Composición: $f(g(x))$.

Para el caso de los desplazamientos verticales y horizontales, tenemos que siendo c una constante mayor que cero:

- Desplazamiento vertical: $f(x) + c$. Desplaza la función c unidades hacia arriba.
- Desplazamiento horizontal: $f(x + c)$. Desplaza la función c unidades hacia la izquierda.

```python
def f(x):
    return x**2

c = 5

N = 100

x = np.linspace(-10, 10, N) 

y = f(x) + c
y2 = f(x) - c
y3 = f(x + c)
y4 = f(x - c)

fig, ax = plt.subplots(2,2, figsize=(18,8))
graph2((0,0),y,'f(x) + c')
graph2((0,1),y2,'f(x) - c')
graph2((1,0),y3,'f(x + c)')
graph2((1,1),y4,'f(x - c)')
plt.tight_layout()
plt.show()
```

Para el caso de los alargamientos y compresiones, tenemos que siendo c una constante mayor que cero:

- Alargamiento vertical: $c \cdot f(x)$. Alarga la función c veces.
- Compresión vertical: $\frac{1}{c} \cdot f(x)$. Comprime la función c veces.
- Alargamiento horizontal: $f(c \cdot x)$. Alarga la función c veces.
- Compresión horizontal: $f(\frac{1}{c} \cdot x)$. Comprime la función c veces.

```python
def f(x):
    return np.sin(x)

c = 5

y = c * f(x)
y2 = 1/c * f(x)
y3 = f(c*x)
y4 = f(x/c)

fig, ax = plt.subplots(2,2, figsize=(18,8))
graph2((0,0),y,'c * f(x)')
graph2((0,1),y2,'1/c * f(x)')
graph2((1,0),y3,'f(c*x)')
graph2((1,1),y4,'f(x/c)')
plt.tight_layout()
plt.show()
```

**Reflexiones**

- Reflexión vertical: $-f(x)$. Refleja la función en el eje $x$.
- Reflexión horizontal: $f(-x)$. Refleja la función en el eje $y$.

```python
N = 1000

def f(x):
    return x**3 + 150

x = np.linspace(-10, 10, N)

y = f(x)
y2 = f(-x)
y3 = -f(x)

fig, ax = plt.subplots(1,3, figsize=(18,8))
graph2((0),y,'f(x)')
graph2((1),y2,'f(-x)')
graph2((2),y3,'-f(x)')
plt.tight_layout()
plt.show()
```

**Perceptrón**

Empezando desde una neurona real, vemos que existen tres partes importantes:
- Dendritas: Son las entradas de la neurona.
- Cuerpo celular: Es el procesador de la neurona.
- Axón: Es la salida de la neurona.

Modelando esto de una manera matemática, tenemos que:



Tenemos entonces diferentes señales de entrada, pesos sinápticos, union sumadora, función de activación y que llevan a una salida.

**Señal de Entrada**

La señal de entrada es la información que recibe la neurona. Esta información puede ser de diferentes tipos, como por ejemplo:
- Señal de entrada continua: $x(t)$.
- Señal de entrada discreta: $x[n]$.
- Señal de entrada binaria: $x[n] \in \{0, 1\}$.


**Peso Sináptico**

El peso sináptico es la importancia que tiene una señal de entrada para la neurona.

**Union Sumadora**

La union sumadora es la suma de las señales de entrada multiplicadas por sus pesos sinápticos.

$\sum_{i=1}^{n} x_i \cdot w_i$

**Función de Activación**

La función de activación es la función que se aplica a la union sumadora para obtener la salida de la neurona. El valor de $f(x)$ se usa para clasificar $x$ como perteneciente a una clase o a otra. 

La u representa el umbral de la neurona. Si la union sumadora es mayor que el umbral, la neurona se activa. Si la union sumadora es menor que el umbral, la neurona no se activa.

Ahora, una sola neurona no tiene sentido, ya que solo puede clasificar dos clases. Para poder clasificar más de dos clases, tenemos que usar una red neuronal. Normalmente, el axón de una neurona se conecta con el cuerpo celular de otra neurona. El perceptón es una red neuronal con una sola capa.

**Regresión Lineal Simple**

Estas trabajando en una compania como Cientifico/a de Datos, llega tu gerente y te da la siguiente informacion.

En el ultimo mes se ha gastado $X$ de publicidad, y obtuvieron $Y$ de retorno en ventas.

Se te pide que elabores un modelo o funcion que te trate de predecir, cuanto se va a vender dependiendo de cuanto se vaya a gastar.

| Publicidad | Ganancias |
|------------|-----------|
| 1.200000	 | 2.000000  |
| 2.000000	 | 3.000000  |
| 3.200000	 | 3.400000  |
| 2.500000	 | 3.100000  |
| 5.000000	 | 4.000000  |
| 6.000000	 | 4.700000  |
| 4.000000	 | 3.800000  |
| 8.000000	 | 7.000000  |


```python
x = np.array([1.200000, 2.000000, 3.200000, 2.500000, 5.000000, 6.000000, 4.000000, 8.000000])
y = np.array([2.000000, 3.000000, 3.400000, 3.100000, 4.000000, 4.700000, 3.800000, 7.000000])
n = len(x)

xy = np.sum((x * y), axis= 0)

x2 = np.sum((x**2), axis= 0)

sum_x = np.sum(x, axis= 0)
sum_y = np.sum(y, axis= 0)

def m():
    return ((n * xy) - (sum_x * sum_y)) / ((n * x2) - (sum_x**2))

m = m()

def b():
    return (sum_y - (m * sum_x)) / n

b = b()

def f():
    return m*x + b

y_prima = f()

df = pd.DataFrame.from_dict({
    "xy"      :xy,
    "x_2"     :x2,
    "sum_x"   :sum_x,
    "sum_y"   :sum_y,
    "m"       :m,
    "b"       :b,
    "y_prima" :y_prima,
},)
```
