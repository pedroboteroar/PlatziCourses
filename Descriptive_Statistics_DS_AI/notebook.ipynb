# Estadística Descriptiva para Ciencia de Datos

**Estadística Descriptiva vs Estadística Inferencial**

Dando como ejemplo las estadísticas de un jugador de fútbol, tenemos que la estadística descriptiva busca resumir el historial deportivo del jugador, mientras que la estadística inferencial busca predecir el rendimiento del jugador en el futuro.

Siendo así, tenemos que:

- La estadística descriptiva se enfoca en describir y resumir los datos, para entender de manera sencilla y concreta toda la información que necesitamos sobre un asunto determinado.
- La estadística inferencial se enfoca en predecir y tomar decisiones sobre un asunto determinado, basándose en los datos que se tienen.

La estadística descriptiva tiene un problema en el sentido de que pueden llegar a mentir de acuerdo a la forma en la que se presentan los datos, por lo que es importante saber interpretarlos.

Volviendo al caso de los jugadores de fútbol, tenemos que la estadística descriptiva nos puede decir quien es el mejor jugador de fútbol, pero esto va a depender de la definición que se le de al mejor jugador. Esto lleva a que no haya una definición objetiva de quien es el mejor jugador de fútbol, por lo que se debe tener cuidado al momento de interpretar los datos..

**Medidas de Tendencia Central**

Las medidas de tendencia central son medidas que nos permiten resumir un conjunto de datos en un solo valor que representa el centro de la distribución de los datos.

Las medidas de tendencia central más comunes son:

- Media: Es el promedio de los datos.

$$\mu = \frac{1}{n} \sum_{i=1}^{n} x_i$$

- Mediana: Es el valor que se encuentra en la mitad de los datos.

$$\tilde{x} = \begin{cases} x_{\frac{n+1}{2}} & \text{si n es impar} \\ \frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2} & \text{si n es par} \end{cases}$$

- Moda: Es el valor que más se repite en los datos.

$$\text{Moda} = \text{Valor que más se repite}$$

Un diagrama de frecuencias es un diagrama que nos permite visualizar la frecuencia de los datos. La frecuencia de los datos es el número de veces que se repite cada valor. Los diagramas de frecuencias se pueden hacer para datos categóricos y para datos numéricos. Para datos categóricos se usan gráficos de barras y para datos numéricos se usan histogramas.

Para construir un diagrama de frecuencias se deben seguir los siguientes pasos:

- Encontrar el número de clases.
- Encontrar el rango de los datos.
- Encontrar el ancho de las clases.

- Encontrar las frecuencias de las clases.
- Encontrar las marcas de clase.
- Encontrar las frecuencias relativas de las clases.
- Encontrar las frecuencias acumuladas de las clases.
- Encontrar las frecuencias relativas acumuladas de las clases.

**Medidas de Dispersión**

Las medidas de dispersión son medidas que nos permiten resumir un conjunto de datos en un solo valor que representa la dispersión de los datos.

Las medidas de dispersión más comunes son:

- Rango: Es la diferencia entre el valor máximo y el valor mínimo.

$$\text{Rango} = \text{Valor Máximo} - \text{Valor Mínimo}$$

- Rango Intercuartílico: Es la diferencia entre el tercer cuartil y el primer cuartil.

$$\text{Rango Intercuartílico} = \text{Tercer Cuartil} - \text{Primer Cuartil}$$

- Varianza: Es el promedio de las diferencias al cuadrado entre cada valor y la media.

$$\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2$$

- Desviación Estándar: Es la raíz cuadrada de la varianza.

$$\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}$$

- Coeficiente de Variación: Es la desviación estándar dividida entre la media.

$$\text{CV} = \frac{\sigma}{\mu}$$



Ahora, un cuartil es un valor que divide a los datos en cuatro partes iguales. El primer cuartil es el valor que divide a los datos en cuatro partes iguales, de tal forma que el 25% de los datos son menores o iguales al primer cuartil. El segundo cuartil es el valor que divide a los datos en cuatro partes iguales, de tal forma que el 50% de los datos son menores o iguales al segundo cuartil. El tercer cuartil es el valor que divide a los datos en cuatro partes iguales, de tal forma que el 75% de los datos son menores o iguales al tercer cuartil.

Una forma efectiva de graficar estas medidas de dispersión es usando un diagrama de caja y bigotes. Un diagrama de caja y bigotes es un diagrama nos permite visualizar las medidas de dispersión de los datos. Un diagrama de caja y bigotes se puede hacer para datos categóricos y para datos numéricos.

Las distribuciones de los datos pueden ser simetricas o asimetricas. Una distribución es simetrica cuando la media, la mediana y la moda son iguales. Una distribución es asimetrica cuando la media, la mediana y la moda son diferentes.

Para identificar si una distribución es simetrica o asimetrica se puede usar un diagrama de caja y bigotes. Si la mediana se encuentra en el centro de la caja, entonces la distribución es simetrica. Si la mediana se encuentra desplazada hacia la parte inferior de la caja, entonces la distribución es asimetrica positiva. Si la mediana se encuentra desplazada hacia la parte superior de la caja, entonces la distribución es asimetrica negativa.

Ahora, la curtosis es una medida que nos permite identificar si una distribución es puntiaguda o achatada. Una distribución es puntiaguda cuando la curtosis es mayor que cero. Una distribución es achatada cuando la curtosis es menor que cero. Una distribución es normal cuando la curtosis es igual a cero. Para distribuciones asimetricas, la curtosis es mayor que cero cuando la distribución es asimetrica positiva y la curtosis es menor que cero cuando la distribución es asimetrica negativa.

Una asimetria es positiva cuando la media es mayor que la mediana. Una asimetria es negativa cuando la media es menor que la mediana. Una asimetria es cero cuando la media es igual que la mediana.

**Escalamiento Lineal**

El escalamiento lineal es una técnica que nos permite transformar los datos para que tengan una media de cero y una desviación estándar de uno. El escalamiento lineal se puede hacer para datos categóricos y para datos numéricos.

Hay diferentes tipos de escalamiento lineal. A continuación, se presentan los más comunes.

- Escalamiento min-max: Es una técnica que nos permite transformar los datos para que tengan un valor mínimo de cero y un valor máximo de uno.

$$x_{\text{esc}} = \frac{2x - \min(x) - \max(x)}{\max(x) - \min(x)}$$

- Clipping: Es una técnica que nos permite transformar los datos para que tengan un valor mínimo de cero y un valor máximo de uno. Si un valor es menor que cero, entonces se reemplaza por cero. Si un valor es mayor que uno, entonces se reemplaza por uno. Fuerza todos los datos por fuera para que estén dentro del rango. NO es la mejor técnica porque puede alterar la distribución de los datos.

$$x_{\text{esc}} = \begin{cases} 0 & \text{si } x < 0 \\ 1 & \text{si } x > 1 \\ x & \text{si } 0 \leq x \leq 1 \end{cases}$$

- Z-Score: Es una técnica que nos permite transformar los datos para que tengan una media de cero y una desviación estándar de uno. Esta consiste en calcular los valores normalizados a partir del promedio y la desviación estándar de los datos.

$$x_{\text{esc}} = \frac{x - \mu}{\sigma}$$

**Escalamiento No Lineal**

El escalamiento no lineal es una técnica que nos permite transformar los datos para que tengan una media de cero y una desviación estándar de uno. El escalamiento no lineal se puede hacer para datos categóricos y para datos numéricos.

Hay diferentes tipos de escalamiento no lineal. A continuación, se presentan los más comunes:

- Logística: Los valores se tranforman usando la función logística. Donde $K$ es el límite del umbral, $C$ es la constante de integración, $r$ es la tasa de crecimiento y $x$ es el valor a transformar.

$$x_{\text{esc}} = \frac{K}{1 + Ce^{-rx}}$$

- Tangente Hiperbólica: Los valores se tranforman usando la función tangente hiperbólica.

$$x_{\text{esc}} = \tanh(x)$$

- Sigmoide: Es una varsión de la función logística. Donde se asume que los límites del umbral son cero y uno.

$$x_{\text{esc}} = \frac{1}{1 + e^{-x}}$$

```python
X, y = datasets.load_diabetes(return_X_y=True)
raw = X[:, None, 2]
```

```python
# Escalamiento Max-Min

max_raw = max(raw)
min_raw = min(raw)

scaled = (2*raw - min_raw + max_raw) / (max_raw - min_raw)

fig, axs = plt.subplots(ncols=2, sharex= True, figsize=(12, 5))
sns.histplot(raw, ax=axs[0])
sns.histplot(scaled, ax=axs[1])
```

```python
# Entrenar el modelo
def train_raw():
    regr = linear_model.LinearRegression()
    regr.fit(raw, y)

# Entrenar el modelo con los datos escalados
def train_scaled():
    regr = linear_model.LinearRegression()
    regr.fit(scaled, y)
```

```python
raw_time = timeit.timeit(train_raw, number=10000)
scaled_time = timeit.timeit(train_scaled, number=10000)

print(f'Raw data training time: {raw_time}')
print(f'Scaled data training time: {scaled_time}')
```

```python
# Z-Score

mean_raw = np.mean(raw)
std_raw = np.std(raw)

z_score = (raw - mean_raw) / std_raw

fig, axs = plt.subplots(ncols=2, sharex= True, figsize=(12, 5))
sns.histplot(raw, ax=axs[0])
sns.histplot(z_score, ax=axs[1])
```

```python
def train_z_score():
    regr = linear_model.LinearRegression()
    regr.fit(z_score, y)

z_score_time = timeit.timeit(train_z_score, number=10000)

print(f'Z-Score data training time: {z_score_time}')
```

**Correlación y Covarianza**

La correlación es una medida que nos permite identificar la relación entre dos variables, dandonos un valor que representa que tanto estan relacionadas de manera lineal dos variables. 

La covarianza es un valor que nos indica el grado de variación conjunta de dos variables aleatorias respecto a sus medias. 

Ahora, el coeficiente de correlación es la medida específica que cuantifica la intensidad de la relación lineal entre dos variables en un análisis de correlación. El coeficiente de correlación es un valor que va de -1 a 1. Un valor de -1 indica una correlación negativa perfecta. Un valor de 1 indica una correlación positiva perfecta. Un valor de 0 indica que no hay correlación entre las variables.



La correlación no implica causalidad.

La matriz de covarianza está dada por:

$$\Sigma = \begin{bmatrix} \sigma_{11} & \sigma_{12} & \cdots & \sigma_{1n} \\ \sigma_{21} & \sigma_{22} & \cdots & \sigma_{2n} \\ \vdots & \vdots & \ddots & \vdots \\ \sigma_{n1} & \sigma_{n2} & \cdots & \sigma_{nn} \end{bmatrix}$$

Donde $\sigma_{ij}$ es la covarianza entre la variable $i$ y la variable $j$. Esto se calcula, a nivel muestral, como:

$$\sigma_{ij} = \frac{1}{n} \sum_{k=1}^n (x_{ik} - \mu_i)(x_{jk} - \mu_j)$$

Teniendo esta matriz, podemos calcular la matriz de correlación:

$$\rho = \begin{bmatrix} \rho_{11} & \rho_{12} & \cdots & \rho_{1n} \\ \rho_{21} & \rho_{22} & \cdots & \rho_{2n} \\ \vdots & \vdots & \ddots & \vdots \\ \rho_{n1} & \rho_{n2} & \cdots & \rho_{nn} \end{bmatrix}$$

Donde $\rho_{ij}$ es el coeficiente de correlación entre la variable $i$ y la variable $j$. Que se calcula como:

$$\rho_{ij} = \frac{\sigma_{ij}}{\sigma_i \sigma_j}$$

**Calculo de valores propios de una matriz**

El que dos variables estén altamente correlacionados implica que aportan básicamente la misma información del problema por lo que podríamos tal vez trabajar con una sola de ellas. Esa reducción de dimensionalidad es lo que se conoce como análisis de componentes principales.

El PCA (Análisis de Componentes Principales) está basado en un concepto del álgebra lineal llamado valores y vectores propios. Los valores propios son los valores de $\lambda$ que satisfacen la ecuación:

$$\Sigma v = \lambda v$$

Donde $\Sigma$ es la matriz de covarianza y $v$ es un vector propio. Los vectores propios son los vectores $v$ que satisfacen la ecuación anterior.

**Breve Repaso de términos de Álgebra Lineal**

Las matrices son arreglos bidimensionales de números que constan de un cierto número de filas y columnas que denominamos dimensiones de la matriz.

Trasnponer una matriz es cambiar las filas por las columnas. Por ejemplo:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}$$

$$A^T = \begin{bmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{bmatrix}$$

Para sumar o restar matrices, estas deben tener las mismas dimensiones. La suma o resta de matrices se realiza sumando o restando los elementos de las matrices que ocupan la misma posición. Por ejemplo:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}$$

$$B = \begin{bmatrix} 7 & 8 & 9 \\ 10 & 11 & 12 \end{bmatrix}$$

$$A + B = \begin{bmatrix} 1 + 7 & 2 + 8 & 3 + 9 \\ 4 + 10 & 5 + 11 & 6 + 12 \end{bmatrix} = \begin{bmatrix} 8 & 10 & 12 \\ 14 & 16 & 18 \end{bmatrix}$$

Para multiplicar matrices entre ellas, el número de columnas de la primera matriz debe ser igual al número de filas de la segunda matriz. El resultado de la multiplicación es una matriz con el número de filas de la primera matriz y el número de columnas de la segunda matriz. Por ejemplo:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}$$

$$B = \begin{bmatrix} 7 & 8 \\ 9 & 10 \\ 11 & 12 \end{bmatrix}$$

$$A \cdot B = \begin{bmatrix} 1 \cdot 7 + 2 \cdot 9 + 3 \cdot 11 & 1 \cdot 8 + 2 \cdot 10 + 3 \cdot 12 \\ 4 \cdot 7 + 5 \cdot 9 + 6 \cdot 11 & 4 \cdot 8 + 5 \cdot 10 + 6 \cdot 12 \end{bmatrix} = \begin{bmatrix} 58 & 64 \\ 139 & 154 \end{bmatrix}$$

```python
# En Python, la función np.matmul() se utiliza para multiplicar matrices.

A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])

np.matmul(A, B)
```

Las matrices pueden tener una matriz inversa. La matriz inversa de una matriz $A$ se denota como $A^{-1}$ y se define como la matriz que cumple la siguiente ecuación:

$$A \cdot A^{-1} = A^{-1} \cdot A = I$$

Donde $I$ es la matriz identidad. La matriz identidad es una matriz cuadrada con unos en la diagonal y ceros en el resto de las posiciones. Por ejemplo:

$$I = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

```python
# Cálculo de matriz inversa

A = np.array([[2, 4], [-1, 2]])
A_inv = np.linalg.inv(A)
A_inv
```

```python
# Comprobando la matriz inversa

np.matmul(A, A_inv)
```

Ahora, el producto de una matriz cuadrada por un vector con longitud igual al numero de filas de la matriz, da como resultado un vector con longitud igual al número de columnas de la matriz. Por ejemplo:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$$

$$v = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}$$

$$A \cdot v = \begin{bmatrix} 1 \cdot 1 + 2 \cdot 2 + 3 \cdot 3 \\ 4 \cdot 1 + 5 \cdot 2 + 6 \cdot 3 \\ 7 \cdot 1 + 8 \cdot 2 + 9 \cdot 3 \end{bmatrix} = \begin{bmatrix} 14 \\ 32 \\ 50 \end{bmatrix}$$

De igual manera, la multiplicación entre dos vectores, donde uno de ellos está transpuesto, da como resultado un valor unico. Por ejemplo:

$$v = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}$$

$$w = \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix}$$

$$v^T \cdot w = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix} \cdot \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} = 1 \cdot 4 + 2 \cdot 5 + 3 \cdot 6 = 32$$

La longitud de un vector se calcula como:

$$\sqrt{v^T \cdot v}$$

Por ejemplo:

$$v = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}$$

$$\sqrt{v^T \cdot v} = \sqrt{\begin{bmatrix} 1 & 2 & 3 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}} = \sqrt{1 \cdot 1 + 2 \cdot 2 + 3 \cdot 3} = \sqrt{14} = 3.741657$$

**Vectores y Valores Propios**

Un vector propio de una matriz cuadrada $A$ es un vector $v$ que cumple la siguiente ecuación:

$$A \cdot v = \lambda \cdot v$$

Aquí, $A$ es una matriz cuadrada NxN y $x$ es un vector columna. 

Tomemos como ejemplo la siguiente matriz:

$$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$$

Para encontrar los vectores propios de esta matriz, debemos resolver la siguiente ecuación:

$$\begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} \cdot \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = \lambda \cdot \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$$

$$\begin{bmatrix} v_1 + 2 \cdot v_2 \\ 3 \cdot v_1 + 4 \cdot v_2 \end{bmatrix} = \lambda \cdot \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$$

$$\begin{bmatrix} v_1 + 2 \cdot v_2 - \lambda \cdot v_1 \\ 3 \cdot v_1 + 4 \cdot v_2 - \lambda \cdot v_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

$$\begin{bmatrix} v_1 \cdot (1 - \lambda) + 2 \cdot v_2 \\ 3 \cdot v_1 + v_2 \cdot (4 - \lambda) \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$
